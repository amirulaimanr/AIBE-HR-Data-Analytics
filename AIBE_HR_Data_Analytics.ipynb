# Build a predictive model

Step #1. Pick the right features / inputs for our ML model

1. What is our target variable ?
2. What are the inputs we should use ?

Answer : 1. Left
         2. Satisfication level, experience, number of project

````python
inputs = ["number_project", "satisfaction_level", "exp_in_company"]
target = 'left'

X = hr_data[inputs]
y = hr_data[target]
````

````python
# We need to split our data into training set and testing set
# 80% of the data to train
# 20% of the data to test
# If you use 100% to train, then we got nothing to test
# so we have to keep 20% to test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
````

# Model selection
1. We need to pick a model
2. Two types of ML 

*  Supervised
*  Unsupervised


3.  Supervised learning 


*   a. predict a value (regression)
*   b. predict an outcome (classification)

4. Classification

*   KNN
*   Logistic regression
*   Decision tree


````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train) # fit the data points into our model
````

````python
# this is when we use the testing set to check the accuracy (performance)
# of our model

from sklearn.metrics import accuracy_score
predictions = knn.predict(X_test)
accuracy_score(y_test, predictions)
````

````python
# This represents a 94.97% accuracy while making predictions

knn.predict([[7, 0.8, 5]])
````
